### 解答

````python
import numpy as np

class Board:
    # 初期化
    def __init__(self, pieces=None, enemy_pieces=None):
        # 石の配置
        def piece_arrange(input_pieces):
#### ここを自分で書く ######
            if input_pieces != None:
                return input_pieces
            return np.zeros(9)
            # 次のコメントに従って，piece_arrange(input_pieces)を完成させなさい．
            # もしinput_piecesがNoneなら, 初期化として0が9個の1次元配列をreturnする(これは石がどこにもないことを意味する)
            # もしinput_piecesがNoneでないなら，input_piecesをそのままreturnする

#### ここまで ######
        self.pieces = piece_arrange(pieces)
        self.enemy_pieces = piece_arrange(enemy_pieces)

    # 石の数の取得
    def piece_count(self, pieces):
#### ここを自分で書く ######
        
        # 石の数 (配列piecesの1の数)を数えcountに与えなさい．
        count: int = sum(pieces)
#### ここまで ######
        return count

    # 合法手のリストの取得
    def legal_actions(self):
        actions = []
#### ここを自分で書く ######
        for i in range(9):
            if self.pieces[i] == self.enemy_pieces[i] == 0:
                actions.append(i)
        # 合法手を探して，そのマス目の番号を配列actionsの中に入れなさい．
        # 配列に要素を足すときはappendを使う．
        # 自分と敵の石がないところを探して，actionsにappendする．

#### ここまで ######
        return actions
        
    # 先手かどうか
    def is_first_player(self):
#### ここを自分で書く ######
        if self.piece_count(self.pieces) == self.piece_count(self.enemy_pieces):
            return True
        return False
        # 先手の場合にTRUEを返すメソッドを与えなさい．
        # 先手の場合は，先手と後手の石の数が等しいことを条件に使える

#### ここまで ######
````

````python
class State(Judge):
    def __init__(self, pieces=None, enemy_pieces=None):
        Board.__init__(self, pieces, enemy_pieces)

    # ゲーム終了かどうか
    def is_done(self):
        return self.is_lose() or self.is_draw()

    # 次の状態の取得
    def next(self, action):
        pieces = self.pieces.copy() # piecesに変更があっても，別のpiecesには影響がないコピーの仕方
#### ここを自分で書く ######
        pieces[action] = 1
        # actionは，合法手の中から１つ，石を置ける位置を0~8までの整数で与える．
        # actionにしたがって石を置きなさい．piecesの特定の場所の変数を1に変えれば良い．

#### ここまで ######
        return State(self.enemy_pieces, pieces) # enemy_piecesとpiecesの入れ替えに注意．

    # 文字列表示
    def __str__(self):
        ox = ('O', 'X') if self.is_first_player() else ('X', 'O')
        str = ''
        for i in range(9):
            if self.pieces[i] == 1:
                str += ox[0]
            elif self.enemy_pieces[i] == 1:
                str += ox[1]
            else:
                str += '-'
            if i % 3 == 2:
                str += '\n'
        return str
````
